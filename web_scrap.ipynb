```python
def get_resto_dishes_from_link(website_link):
    
    print("called")
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36', "Upgrade-Insecure-Requests": "1","DNT": "1","Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8","Accept-Language": "en-US,en;q=0.5","Accept-Encoding": "gzip, deflate"}
    response = requests.get(website_link, headers=headers)
    resto_dishes_soup = BeautifulSoup(response.content, "html5lib")
    main_detail_div = resto_dishes_soup.body
    maindiv = main_detail_div.find("main").findChild()
    sections = maindiv.findChildren("section", recursive=False)
    section_3 = sections[3]
    parent_detail_div = section_3.findChild().findChildren("section", recursive=False)[1]
    dishes_sections = parent_detail_div.findChildren("section", recursive=False)
    resto_dishes_list = []
    for index in range(1, len(dishes_sections)):
        main_inner_divs = dishes_sections[index].findChildren("div", recursive=False)
        for main_inner_div in main_inner_divs:
            inner_divs = main_inner_div.findChildren("div", recursive=False)
            for inner_div in inner_divs:
                item_div = inner_div.findChild().findChild().findChildren("div", recursive=False)[1]
                dish_list_item = item_div.findChildren("div", recursive=False)[0].findChildren("div", recursive=False)[0]
                resto_dishes_list.append(dish_list_item.find_all("h4")[0].contents[0])
    resto_dishes_list
```

```python
soup = BeautifulSoup (open("zomato_restos_new.htm"), features="lxml")
mydivs = soup.body.find_all("div", {"class": "sc-1mo3ldo-0 sc-LKuAh dRasFA"})
```

```python
def get_resto_name_link_from_list_of_divs(mydivs):
    restos = {}
    for restoList in mydivs:
        list1 = restoList.findChild()
        children = list1.findChildren("div", recursive=False)
        for rest in children:
            jumbo_tracker = rest.findChild()
            jumbo_child = jumbo_tracker.findChild()
            hyperlink = jumbo_child.findChildren("a", recursive=False)[1]
            #print(hyperlink.get('href'))
            website_link = hyperlink.get('href')
            #print(website_link)
            title = hyperlink.findChildren("p", recursive=False)[0].contents[0]
            #cusines = hyperlink.findChildren("p", recursive=False)[1].contents[0]
            now = datetime.datetime.now()
            date_time = now.strftime("%H:%M:%S")
            #print(date_time)
#            time.sleep(10)
            print(title)
            headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36', "Upgrade-Insecure-Requests": "1","DNT": "1","Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8","Accept-Language": "en-US,en;q=0.5","Accept-Encoding": "gzip, deflate"}
            response = requests.get(website_link, headers=headers)
            resto_dishes_soup = BeautifulSoup(response.content, "html5lib")
            main_detail_div = resto_dishes_soup.body
            try:
                maindiv = main_detail_div.find("main").findChild()
                sections = maindiv.findChildren("section", recursive=False)
                section_3 = sections[3]
                parent_detail_sections = section_3.findChild().findChildren("section", recursive=False)
                if len(parent_detail_sections) > 0:
                    parent_detail_div = parent_detail_sections[1]
                    dishes_sections = parent_detail_div.findChildren("section", recursive=False)
                    resto_dishes_list = []
                    for index in range(1, len(dishes_sections)):
                        main_inner_divs = dishes_sections[index].findChildren("div", recursive=False)
                        for main_inner_div in main_inner_divs:
                            inner_divs = main_inner_div.findChildren("div", recursive=False)
                            for inner_div in inner_divs:
                                item_div = inner_div.findChild().findChild().findChildren("div", recursive=False)[1]
                                dish_list_item = item_div.findChildren("div", recursive=False)[0].findChildren("div", recursive=False)[0]
                                resto_dishes_list.append(dish_list_item.find_all("h4")[0].contents[0])
#                    print(resto_dishes_list)
                    restos[title] = resto_dishes_list
            except:
                print("An exception occurred")
                return restos
            #df= df.append({'name':title,'dishes':resto_dishes_list})
            #resto_dishes = get_resto_name_link_from_list_of_divs(website_link)
            #df= df.append({'name':title,'dishes':resto_dishes})
    return restos
```
